```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
all_data= pd.read_csv('etf_options.csv')
all_data=all_data[all_data['DataDate']<='2021-04-14']
all_data=all_data[all_data['DataDate']>='2016-11-01']
all_data['DataDate'] = all_data.DataDate.apply(lambda x: int(x.replace("-", "")))
```

```python
all_data['year'] = all_data['DataDate'].apply(lambda x: x // 10000)
```

```python
s = all_data[['year', 'DataDate']].drop_duplicates().reset_index(drop=True).pct_change().dropna()
s[s['year']!=0]
```
